```python
import pandas as pd
import numpy as np 
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv("tmp_trimmed_df.csv")
df
```

```python
columns = df.columns[-5:]
df[columns[1]] = df[columns[1]].shift(periods=-1, fill_value="0")
df[columns[2]] = df[columns[2]].shift(periods=-2, fill_value="0")
df[columns[3]] = df[columns[3]].shift(periods=-3, fill_value="0")
df[columns[4]] = df[columns[4]].shift(periods=-4, fill_value="0")
df = df.iloc[::5, 1:].reset_index(drop=True)
df
```
